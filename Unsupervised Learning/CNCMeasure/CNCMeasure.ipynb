# 3. Model Development

```python
from sklearn.mixture import GaussianMixture

# Fit Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)  # 4 clusters
gmm.fit(X)
labels = gmm.predict(X)

# Plot the GMM clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.6)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.title('CNC Measurements')
plt.xlabel('Dimension 1 (mm)')
plt.ylabel('Dimension 2 (mm)')
plt.legend()
plt.grid()
plt.show()
```
