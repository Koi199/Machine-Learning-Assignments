```python
import seaborn as sns

# Assuming you've loaded your dataset into a DataFrame called 'df'
def plot_feature_distributions(df, state_column='State'):
    # Get unique states
    states = df[state_column].unique()
    
    # Select numerical columns (exclude the state column)
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    numerical_columns = [col for col in numerical_columns if col != state_column]
    
    # Create a figure with subplots
    n_cols = 3
    n_rows = (len(numerical_columns) + n_cols - 1) // n_cols
    
    plt.figure(figsize=(20, 4*n_rows))
    
    # Plot each feature
    for i, feature in enumerate(numerical_columns, 1):
        plt.subplot(n_rows, n_cols, i)
        
        # Box plot for the feature across different states
        sns.boxplot(x=state_column, y=feature, data=df)
        
        plt.title(f'{feature} Distribution')
        plt.xticks(rotation=45)
        plt.tight_layout()
    
    plt.show()

# Heatmap of feature correlations for each state
def plot_state_correlation_heatmaps(df, state_column='State'):
    # Get unique states
    states = df[state_column].unique()
    
    # Select numerical columns
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    numerical_columns = [col for col in numerical_columns if col != state_column]
    
    # Create a figure
    plt.figure(figsize=(20, 5*len(states)))
    
    # Plot correlation heatmap for each state
    for i, state in enumerate(states, 1):
        plt.subplot(len(states), 1, i)
        
        # Filter data for the current state
        state_data = df[df[state_column] == state][numerical_columns]
        
        # Compute correlation matrix
        corr_matrix = state_data.corr()
        
        # Plot heatmap
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)
        plt.title(f'Correlation Heatmap for {state} State')
        plt.tight_layout()
    
    plt.show()

# Mean feature values across states
def plot_mean_features_by_state(df, state_column='State'):
    # Select numerical columns
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    numerical_columns = [col for col in numerical_columns if col != state_column]
    
    # Group by state and calculate mean
    state_means = df.groupby(state_column)[numerical_columns].mean()
    
    # Transpose for easier plotting
    state_means_transposed = state_means.T
    
    # Plot
    plt.figure(figsize=(15, 10))
    state_means_transposed.plot(kind='bar', ax=plt.gca())
    plt.title('Mean Feature Values Across States')
    plt.xlabel('Features')
    plt.ylabel('Mean Value')
    plt.legend(title='States', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.xticks(rotation=90)
    plt.grid(True)
    plt.show()

# Call the plotting functions
plot_feature_distributions(ready_dataset)
plot_state_correlation_heatmaps(ready_dataset)
plot_mean_features_by_state(ready_dataset)
```

```python
#Creating Synthetic Data
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats import multivariate_normal
import warnings

def generate_synthetic_data(df, n_samples=None, method='mixed', random_state=42):
    """
    Generate synthetic data from an existing DataFrame
    
    Parameters:
    -----------
    df : pandas DataFrame
        Original dataframe to generate synthetic data from
    n_samples : int, optional
        Number of synthetic samples to generate. 
        If None, generates same number of samples as original dataset
    method : str, optional
        Synthetic data generation method
        Options: 
        - 'mixed': Combines multiple techniques
        - 'gaussian_copula': Advanced probabilistic method
        - 'multivariate_normal': Simple multivariate normal distribution
        - 'bootstrap': Resampling with noise
    random_state : int, optional
        Seed for reproducibility
    
    Returns:
    --------
    pandas DataFrame
        Synthetic data matching original DataFrame structure
    """
    # Suppress warnings
    warnings.filterwarnings('ignore')
    
    # Set random seed
    np.random.seed(random_state)
    
    # Default to original sample size if not specified
    if n_samples is None:
        n_samples = len(df)
    
    # Separate numeric and categorical columns
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    
    # Create a copy of the original DataFrame
    synthetic_df = df.copy()
    
    if method == 'multivariate_normal':
        # Multivariate normal distribution for numeric columns
        if len(numeric_cols) > 0:
            # Standardize numeric data
            scaler = StandardScaler()
            scaled_numeric = scaler.fit_transform(df[numeric_cols])
            
            # Compute mean and covariance
            mean = np.mean(scaled_numeric, axis=0)
            cov = np.cov(scaled_numeric.T)
            
            # Generate synthetic numeric data
            synthetic_numeric = multivariate_normal.rvs(mean=mean, cov=cov, size=n_samples)
            
            # Inverse transform to original scale
            synthetic_numeric = scaler.inverse_transform(synthetic_numeric)
            
            synthetic_df[numeric_cols] = synthetic_numeric
        
        # Handle categorical columns
        for col in categorical_cols:
            synthetic_df[col] = df[col].sample(n=n_samples, replace=True).values
        
        return synthetic_df
    
    elif method == 'bootstrap':
        # Bootstrap resampling with added noise
        synthetic_df = df.sample(n=n_samples, replace=True)
        
        # Add small gaussian noise to numeric columns
        for col in numeric_cols:
            noise = np.random.normal(0, df[col].std() * 0.1, n_samples)
            synthetic_df[col] += noise
        
        return synthetic_df
    
    elif method == 'mixed':
        # Combine multiple techniques
        # 50% direct resampling, 50% with noise
        resample_count = n_samples // 2
        noise_count = n_samples - resample_count
        
        # Direct resampling
        resampled = df.sample(n=resample_count, replace=True)
        
        # Add noise to resampled data
        for col in numeric_cols:
            noise = np.random.normal(0, df[col].std() * 0.05, resample_count)
            resampled[col] += noise
        
        # Additional samples from multivariate normal
        if len(numeric_cols) > 0:
            scaler = StandardScaler()
            scaled_numeric = scaler.fit_transform(df[numeric_cols])
            mean = np.mean(scaled_numeric, axis=0)
            cov = np.cov(scaled_numeric.T)
            
            synthetic_numeric = multivariate_normal.rvs(mean=mean, cov=cov, size=noise_count)
            synthetic_numeric = scaler.inverse_transform(synthetic_numeric)
            
            noisy_df = pd.DataFrame(synthetic_numeric, columns=numeric_cols)
            
            # Sample categorical columns
            for col in categorical_cols:
                noisy_df[col] = df[col].sample(n=noise_count, replace=True).values
            
        # Combine resampled and synthetic data
        synthetic_df = pd.concat([resampled, noisy_df], ignore_index=True)
        
        return synthetic_df
    
    else:
        raise ValueError("Invalid method. Choose 'mixed', 'multivariate_normal', or 'bootstrap'.")

synthetic_data = generate_synthetic_data(df_filtered, n_samples=133, method='mixed')
df_combined = pd.concat([df_filtered, synthetic_data], ignore_index=True)

df_combined.describe()
```

### Define Plot Confusion Matrix Function

```python
from sklearn.metrics import confusion_matrix

# Function to plot a confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(3, 2))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
```

```python
Muse2TouchANN = tf.keras.models.Sequential([
    # Reduce first layer complexity
    tf.keras.layers.Dense(
        min(X.shape[1] * 2, 64),  # Adaptive layer size
        activation='relu', 
        kernel_regularizer=tf.keras.regularizers.l2(0.001),
        input_shape=(X.shape[1],)
    ),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    
    tf.keras.layers.Dense(
        max(15, X.shape[1]),  # Adaptive second layer
        activation='relu', 
        kernel_regularizer=tf.keras.regularizers.l2(0.001)
    ),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    
    tf.keras.layers.Dense(3, activation='softmax')
])

# Compile with learning rate scheduling
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=10,
    decay_rate=0.9
)
optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

Muse2TouchANN.compile(
    optimizer=optimizer, 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
```

### Fit ANN and Plot ANN Learning Curves Function

```python
def fit_ANN_and_plot_LC(model):
    history = model.fit(X_train, y_train, 
    batch_size=1,  # With 27 samples, this is okay
    epochs=50,  # Increased, but with early stopping
    validation_split=0.2, 
    verbose=1)

    fig, ax = plt.subplots(figsize=(8, 5))  # Create a single plot

    # Plot the loss for the single model
    ax.plot(history.history['loss'], label='Train')
    ax.plot(history.history['val_loss'], label='Validation')

    # Set titles and labels
    ax.set_title("Model Loss Function")
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')

    # Add a legend
    ax.legend(loc="upper left")

    # Show the plot
    plt.show()
```

### Fitting Models

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression# Initialize the classifiers

models = { 
    "Support Vector Machine": SVC(kernel = 'rbf', random_state = 42, probability=True),
    "KNN": KNeighborsClassifier(
    n_neighbors=3,  # With 27 samples, 5 might be too many - try 3
    weights='distance',  # Weight closer neighbors more heavily
    metric='minkowski',
    p=2  # This makes it Euclidean distance
),
    "Random Forest Classification": RandomForestClassifier(
    n_estimators=100,  # Default is good
    max_depth=None,  # Consider limiting depth to prevent overfitting
    min_samples_split=2,
    min_samples_leaf=1,
    class_weight='balanced',
    random_state=42
),
    "xgb": xgb.XGBClassifier(objective ='multi:softmax', num_class=3, random_state=42),
    "Logistic Regression": LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    class_weight='balanced',
    max_iter=1000,
    random_state=42
),
    "ANN": Muse2TouchANN
}

# Dictionary to store accuracy results
accuracy_results = {}

# Train and evaluate each model
for model_name, model in models.items():
    if model_name == "ANN":
            # Fit the ANN model and plot learning curves
            fit_ANN_and_plot_LC(model)
            
            # Make predictions
            y_train_pred = np.argmax(model.predict(X_train), axis=-1)
            y_test_pred = np.argmax(model.predict(X_test), axis=-1)
    else:
            # Fit the other models
            model.fit(X_train, y_train)
            y_train_pred = model.predict(X_train)
            y_test_pred = model.predict(X_test)    
    
    # Calculate the training and test accuracy
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    # Store the results in the dictionary
    accuracy_results[model_name] = {
        "Train Accuracy": train_accuracy * 100,
        "Test Accuracy": test_accuracy * 100
    }
    
    # Print the accuracy for each model
    print(f"{model_name}:")
    print(f"  Training Accuracy: {train_accuracy * 100:.2f}%")
    print(f"  Test Accuracy: {test_accuracy * 100:.2f}%\n")
    
    # Plot confusion matrix for training data
    plot_confusion_matrix(y_train, y_train_pred, model_name+ ' '+'Confusion Matrix (Training Set)')

    # Plot confusion matrix for test data
    plot_confusion_matrix(y_test, y_test_pred, model_name+ ' '+'Confusion Matrix (Test Set)')

# Display accuracy results in a table
import pandas as pd
accuracy_df = pd.DataFrame(accuracy_results).T

accuracy_df
```
